# 자연어 태그 기반 착장 추천 — 데모

자연어로 원하는 분위기를 입력하면, 의미적으로 가장 가까운 착장을 추천한다.

- 임베딩: **BGE-M3** (첫 실행 시 가중치 ~2.2GB 자동 다운로드)
- 데이터: `../data/dataset.zip` (착장 135개 × 7축 태그)
- 스코어: 착장별 상위 3개 축 cosine 평균 (**top-3 mean**)

```python
# 최초 1회만 (이미 설치돼 있으면 생략 가능)
# !pip install -r requirements.txt
```

```python
import recommend as R
from IPython.display import display

# 오프라인 임베딩 빌드(또는 캐시 로드). 첫 실행은 모델 다운로드+인코딩으로 시간이 걸린다.
vectors, meta = R.build_or_load_embeddings()
print(f"착장 {vectors.shape[0]}개 × 축 {vectors.shape[1]}개 × {vectors.shape[2]}차원 임베딩 준비 완료")
```

```python
def show_recommendations(query, top_k=5, size=240):
    print(f"\U0001F50D 쿼리: {query}\n")
    results = R.recommend(query, top_k=top_k)
    for m in results:
        axes = ", ".join(f"{name}({s:.3f})" for name, s in m.matched_axes)
        print(f"[{m.rank}] {m.image_id}  |  score={m.score:.4f}  |  근거 축: {axes}")
        try:
            img = R.load_image(m.image_id)
            display(img.resize((size, int(size * img.height / img.width))))
        except FileNotFoundError as e:
            print(f"   (이미지 표시 불가: {e})")
        print()
```

```python
show_recommendations("오늘 학교 갈 때 입을 깔끔한 꾸안꾸 느낌 추천해줘")
```

```python
show_recommendations("비 오는 날 데이트룩, 차분하고 단정하게")
```

```python
show_recommendations("여름 휴양지에서 시원하고 편한 와이드핏")
```
